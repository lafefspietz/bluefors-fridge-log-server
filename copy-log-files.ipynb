[Jupyter notebook](https://jupyter.org/) to copy log files from log folders from some source path to a local path in the web server directory. For use on [local web servers](https://github.com/lafefspietz/local-web-server) across a local network.

```python
import os
import shutil
```

```python
sourcefolder = 'O:/67203/OnWafMea/DR/Data/2025_05_01_Z_Positioners/Cooldown Logs/'
allfolders = os.listdir(sourcefolder)
logfolders = []
for folder in allfolders:
    if folder[2] == '-' and folder[5] == '-':
        logfolders.append(folder)
logfolders
```

```python
numfolders = 10 # number of folders to fetch from most recent going back
for folder in logfolders[len(logfolders) - numfolders - 1:len(logfolders)]:
    if not os.path.exists('logfolders/' + folder):
        os.mkdir('logfolders/' + folder)    
    logfiles = os.listdir(sourcefolder + folder)
    for logfile in logfiles:
        shutil.copy(sourcefolder + folder + '/' + logfile,'logfolders/' + folder + '/' + logfile)
```
